```python
import pandas as pd
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

print("Matches Dataset:")
display(matches.head())

print("\nDeliveries Dataset:")
display(deliveries.head())

print("\nMatches Info:")
matches.info()

print("\nDeliveries Info:")
deliveries.info()

print("\nMatches Describe:")
display(matches.describe())

print("\nDeliveries Describe:")
display(deliveries.describe())

print("\nMissing values in Matches:")
print(matches.isnull().sum())

print("\nMissing values in Deliveries:")
print(deliveries.isnull().sum())
```

```python
# filling the missing values
matches['city'].fillna('Unknown', inplace=True)
matches['winner'].fillna('No Result', inplace=True)

# dropping the remaining rows with missing values
matches.dropna(inplace=True)
```

```python
deliveries.dropna(inplace=True)
```

```python
#replacing order of the teams with remaining ones
team_name_corrections = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Kings XI Punjab': 'Punjab Kings'
}

matches['team1'] = matches['team1'].replace(team_name_corrections)
matches['team2'] = matches['team2'].replace(team_name_corrections)
matches['winner'] = matches['winner'].replace(team_name_corrections)
matches['toss_winner'] = matches['toss_winner'].replace(team_name_corrections)

# applying same corrections to deliveries dataset
deliveries['batting_team'] = deliveries['batting_team'].replace(team_name_corrections)
deliveries['bowling_team'] = deliveries['bowling_team'].replace(team_name_corrections)
```

```python
# converting date to datetime format
matches['date'] = pd.to_datetime(matches['date'])

# checking again
matches.info()
```

```python
print("Columns in matches.csv:\n")
print(matches.columns.tolist())

# checking a few unique values for categorical columns
print("\nUnique seasons:", matches['season'].unique())
print("\nUnique teams:", matches['team1'].unique())
print("\nUnique venues:", matches['venue'].unique())
```

```python
match_level_features = [
    'id', 'season', 'city', 'date', 'team1', 'team2', 'toss_winner', 'toss_decision',
    'winner', 'result', 'dl_applied', 'win_by_runs', 'win_by_wickets',
    'player_of_match', 'venue', 'umpire1', 'umpire2'
]

print("\nMatch-Level Features:")
for col in match_level_features:
    if col in matches.columns:
        print(f"- {col}")
```

```python
print("\nColumns in deliveries.csv:\n")
print(deliveries.columns.tolist())

# Show unique values for batting_team and bowler
print("\nUnique batting teams:", deliveries['batting_team'].unique())
print("Unique bowlers (sample 5):", deliveries['bowler'].unique()[:5])
```

```python
ball_by_ball_features = [
    'match_id', 'inning', 'batting_team', 'bowling_team', 'over', 'ball',
    'batsman', 'non_striker', 'bowler', 'is_super_over', 'wide_runs', 'bye_runs',
    'legbye_runs', 'noball_runs', 'penalty_runs', 'batsman_runs',
    'extra_runs', 'total_runs', 'player_dismissed', 'dismissal_kind', 'fielder'
]

print("\nBall-by-Ball Features:")
for col in ball_by_ball_features:
    if col in deliveries.columns:
        print(f"- {col}")
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Total Matches Per Season
plt.figure(figsize=(10,5))
sns.countplot(x='season', data=matches, palette='Set2')
plt.title('Total Matches Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# most successful teams(total wins)
top_teams = matches['winner'].value_counts()
plt.figure(figsize=(12,5))
sns.barplot(x=top_teams.index, y=top_teams.values, palette='coolwarm')
plt.title('Most Successful Teams (Total Wins)')
plt.ylabel('Wins')
plt.xlabel('Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#impact of toss
toss_match = matches[matches['toss_winner'] == matches['winner']]
toss_win_rate = len(toss_match) / len(matches)

print(f"Toss winner also won the match in {toss_win_rate:.2%} cases.")

plt.figure(figsize=(6,6))
labels = ['Toss + Match Win', 'Toss ≠ Match Win']
sizes = [len(toss_match), len(matches) - len(toss_match)]
colors = ['#66b3ff','#ff9999']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Toss Impact on Match Outcome')
plt.tight_layout()
plt.show()
```

```python
#top 10 players in ipl history
top_players = matches['player_of_match'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_players.values, y=top_players.index, palette='magma')
plt.title('Top 10 Player of the Match Winners')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.tight_layout()
plt.show()
```

```python
# top 10 run scorers
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette='viridis')
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.tight_layout()
plt.show()
```

```python
#most frequently used stadiums in the ipl history
venue_counts = matches['venue'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette='cubehelix')
plt.title('Top 10 IPL Venues by Match Count')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.tight_layout()
plt.show()
```

```python
#teams wins depending upon batting order
batting_first_wins = matches[matches['result'] == 'runs']['winner'].value_counts()
batting_second_wins = matches[matches['result'] == 'wickets']['winner'].value_counts()
batting_comparison = pd.DataFrame({
    'Batting First Wins': batting_first_wins,
    'Batting Second Wins': batting_second_wins
}).fillna(0)
batting_comparison.plot(kind='bar', figsize=(12,6), stacked=True, colormap='tab20c')
plt.title('Team Wins: Batting First vs Batting Second')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
